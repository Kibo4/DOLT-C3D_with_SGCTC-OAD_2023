```python
import matplotlib.pyplot as plt
import numpy as np 
pathDB = "C:\workspace2\Datasets\MSRC6_IconicC4\\"
pathExpe = pathDB + "expOut/MSRC6_IconicC4/"
```

```python
f = open(pathExpe + "ResultsMultiFold/BoundBasedEval/DetectionToActionPointMF.txt", "r")
data = f.readlines()
f.close()
```

```python
line = list(filter(lambda x: x.startswith("fmeasurePerFrame,"), data))[0]
# get the index of the line
index = data.index(line)
fmeasurePerFrame = list(map(lambda x : float(x)*100, data[index + 1].split(";")))
#split by ";"
fmeasurePerFrame
```

```python
boulahiaResult = np.array([0.417, 0.427, 0.433, 0.453, 0.469, 0.492, 0.524, 0.544, 0.565, 0.594, 0.612, 0.638, 0.652, 0.669,
                  0.684, 0.697, 0.716, 0.73, 0.748, 0.758, 0.777])*100
plt.rcParams.update({'font.size': 18})
# set the line width of plot higher
plt.rcParams['lines.linewidth'] = 3
plt.figure(figsize=(10, 5.5), dpi=180)

print(fmeasurePerFrame[-21:])
absci = range(-20, 1)
plt.plot(absci, fmeasurePerFrame[-21:], color="red")
plt.plot(absci, boulahiaResult, color="blue")
plt.grid(linewidth=0.25)
plt.xlabel('Distance to action point')
plt.xticks(range(-20, 1, 5))
plt.ylabel('Fscore')
plt.legend(["Ours", "ECuDi3D"])

plt.savefig("../Illustrations/ScoreGraphics/EarlyRecognitionBoulahia-Comparison", bbox_inches='tight')
plt.savefig("../Illustrations/ScoreGraphics/EarlyRecognitionBoulahia-Comparison.eps", bbox_inches='tight', format='eps')
```
